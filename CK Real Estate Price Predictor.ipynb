```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(housing, test_size=0.2, random_state = 42)
print(f"Rows in test set: {len(train_set)}\nRows in train set: {len(test_set)}")
```

```python
# there is in CHAS also a some value "1" if "1" is not present in train set then while testing it will give error so that we are splitting it in train & test sets with using StratifiedShuffleSplit.
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing['CHAS']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```
